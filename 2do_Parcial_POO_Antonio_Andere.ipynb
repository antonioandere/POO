**Ejercicio 4:** A partir de un diccionario, crea un DataFrame (que llamarás **precios_muestra**) que consista de dos columnas y que tenga 12 valores **flotantes** que eligirás de manera aleatoria en el intervalo (8,13). Etiqueta del 1 al 12 y llama a la columna de etiquetas Mes.

1.- Construye una función que calcule los rendimientos simples de cada columna.

2.- Las desviaciones estadísticas de los rendimientos ayudan a calcular la volatilidad del portafolio:
$$\text{desviaciones}=(\text{rendimientos}-\text{media}(\text{rendimientos}))^2.$$
Crea una función que calcule la media de las desviaciones y luego la volatilidad de tu portafolio (raíz cuadrada de la media de las desviaciones).

3.- Anualizamos la volatilidad escalándola (multiplicándola) por la raíz cuadrada del número de períodos por observación.

Por lo tanto, para anualizar la volatilidad de una serie mensual, la multiplicamos por la raíz cuadrada de 12. Muestra la serie de la volatilidad anualizada.

4.- Realiza una gráfica de barras para los rendimientos y una gráfica en líneas para los precios.

5.- Calcula los rendimientos compuestos para cada columna y después anualiza los rendimientos.

---

```python
import pandas as pd
import numpy as np

# Creamos un DataFrame con los valores aleatorios
datos = {"Mes": range(1, 13), "Precio": [np.random.uniform(8, 13) for _ in range(1,13)]}
precios_aleatorios= pd.DataFrame(datos)

# Calcular los rendimientos simples de los valores aleatorios, utilizamos el cambio en porcentaje de cada precio.
def calculo_rendimientos(dataframe):
    dataframe["Rendimiento"] = dataframe["Precio"].pct_change()
    return dataframe

precios_muestra = calculo_rendimientos(precios_aleatorios)

# Calculamos la volatilidad anualizada de los rendimientos de los valores aleatorios, obteniendo la media de los rendimientos y obteniendo la desviación estándar.

def calcular_volatilidad_anual(dataframe):
    promedio_rendimientos = dataframe["Rendimiento"].mean()
    desviaciones = (dataframe["Rendimiento"] - promedio_rendimientos)**2
    volatilidad = np.sqrt(desviaciones.mean()) * np.sqrt(12)
    return volatilidad

volatilidad_anualizada = calcular_volatilidad_anual(precios_muestra)

# Realizamos una gráfica que muestre los rendimientos mensuales usando los precios aleatorios.
precios_muestra.set_index("Mes")["Rendimiento"].plot(title="Rendimientos")

# Función para calcular los rendimientos compuestos y anualizarlos
def calcular_rendimientos_compuestos(dataframe):
    rendimiento_compuesto = (1 + dataframe["Rendimiento"]).cumprod()
    rendimiento_anualizados = rendimiento_compuesto[1] ** (12 / len(dataframe))
    return rendimiento_compuesto, rendimiento_anualizados

rendimiento_compuesto, rendimiento_anualizados = calcular_rendimientos_compuestos(precios_muestra)

# Imprimir la volatilidad anualizada
print(volatilidad_anualizada)

# Imprimir los rendimientos anualizados
print(rendimiento_anualizados)

# Mostrar la gráfica de líneas para los precios
precios_muestra.set_index("Mes")["Precio"].plot(title="Precios")
```

**Ejercicio 5:** Carga dos tickers y realiza los pasos del ejercicio anterior usando los tickers que elijas. Recuerda tomar en cuenta la ventana de tiempo que manejarás. Realiza un análisis estadístico (estadística descriptiva, histogramas y gráficas de series de tiempo).

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Descargamos los datos necesarios de los dos tickers, que son AAPL y AMZN.
tickers = ["AAPL", "AMZN"]
fechainicio = "2022-01-01"
fechafin = "2023-01-01"
datos = yf.download(tickers, start=fechainicio, end=fechafin)["Adj Close"]

# Calcular los rendimientos diarios de los instrumentos de AMZN y AAPL
def calculo_rendimientos(dataframe):
    return dataframe.pct_change().dropna()

rendimientos = calculo_rendimientos(datos)

# Calculamos la varianza obteniendo primero la desviación estándar y a partir de ella calculamos la varianza o volatilidad.
def calculo_varianza(dataframe):
    desviacionestandar = (dataframe - dataframe.mean()) ** 2
    varianza= np.sqrt(desviacionestandar.mean())
    return varianza

varianza = calculo_varianza(rendimientos)

# Hacemos el calculo para anualizar la volatilidad
varianza_anualizada = varianza * np.sqrt(12)

# Hacemos la gráfica que muestre los rendimientos de cada instrumento.
rendimientos.plot(title="Rendimientos de AAPL y AMZN")
plt.xlabel("Fecha")
plt.ylabel("Rendimientos")
plt.show()

# Graficamos los precios ajustados de cierre de cada uno de los instrumentos en una serie de tiempo.
datos.plot(title="Adj Close de AAPL y AMZN")
plt.xlabel("Fecha")
plt.ylabel("Adj Close")
plt.show()

# Calculamos los rendimientos de cada instrumento compuestos y anualizados
rendimientos_compuestos = (1 + rendimientos).prod()
rendimientos_anualizados = (rendimientos_compuestos ** (1/len(rendimientos) * 12)) - 1

# Hacemos un análisis estadístico de cada instrumento y sus datos, así como un histograma que muestre los rendimientos de los dos instrumentos.
print("Análisis estadístico")
print(rendimientos.describe())

plt.figure(figsize=(12, 4))
for ticker in tickers:
    plt.hist(rendimientos[ticker], bins=50, alpha=0.6, label=ticker)
plt.legend()
plt.title("Histograma de los rendimientos")
plt.xlabel("Rendimientos")
plt.ylabel("Frecuencia")
plt.show()

# Mostrar la volatilidad anualizada
print(varianza_anualizada)

# Mostrar el rendimiento anualizado
print(rendimiento_anualizados)
```

```python
import numpy as np
import pandas as pd
```

```python
# Creamos la lista de datos a partir de la cual se va a crear el diccionario.
minutosentregoles1=(10,12,32,3.2,4.3)
minutosentregoles2=(12,34,32,2.1,2.2)
minutosentregoles3=(26,24,45,2.1,4.2)
```

```python
#1 Creamos el diccionario con las listas creadas previamente. Le damos un título a cada lista creada.
Dictionary={"Ronaldo":minutosentregoles1,"Neymar":minutosentregoles2,"Messi":minutosentregoles3}
```

```python
Dictionary
```

```python
#2 Aquí podemos obervar el tipo de dato que se obtiene del diccionario creado antes.
type(Dictionary)
```

```python
#3 A partir del diccionario que habíamos creado con las listas anteriores, ahora creamos una base de datos (dataframe) y le realizamos tres segregaciones.
#En las segregaciones, removimos la columna de Neymar y también removimos el renglón 0 y el renglón 4.

dataframe=pd.DataFrame(Dictionary)
dataframe
```

```python
dataframe_segre=dataframe.drop(columns="Neymar")
dataframe_segre
```

```python
dataframe_segre=dataframe_segre.drop(0)
dataframe_segre
```

```python
dataframe_segre=dataframe_segre.drop(4)
dataframe_segre
```

```python
#4 Realizamos un análisis descriptivo del dataframe que se obtuvo después de las segregaciones realizadas. Se observa el número de datos, la media de ambas columnas,
#la desviación estándar, el dato menor de las columnas, el dato máximo de la columna.
dataframe_segre.describe()
```

```python
#5 Realizamos dos filtros de búsqueda sobre el dataframe creado anteriormente, en uno pedimos las columnas done "Messi" tenga un dato mayor a 20 y en
# el segundo pedimos las columnas donde "Ronaldo" tenga un dato menor a 10.
mayora20=dataframe_segre["Messi"]>20
filtrodataframe=dataframe_segre[mayora20]
filtrodataframe
```
